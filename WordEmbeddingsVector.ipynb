```
import gensim
import re
import numpy as np
from gensim.models.keyedvectors import KeyedVectors
```

```
# load pretrained model
W2V_model_CBOW = KeyedVectors.load_word2vec_format('Word2VecCBOW300D5.bin', binary=True)
```

```
# get the top 10 similar words 
word = "كورونا"

most_similar = W2V_model_CBOW.most_similar( word, topn=10 )
for term, score in most_similar:
  if term != word:
    print(term, score)
```

```
## get a word vector
word_vector = W2V_model_CBOW[ word ]
print (word_vector)
```
